```python
import pandas as pd
data = pd.read_excel('ClimateChange.xlsx',sheet_name='Data')  
country = pd.read_excel('ClimateChange.xlsx',sheet_name='Country')
```

```python
data = data[data['Series code']=='EN.ATM.CO2E.KT'].set_index('Country code')
```

```python
data = data.drop(labels=['Country name','Series code','Series name','SCALE','Decimals'],axis=1)
```

```python
data = data.replace({'..':pd.np.NaN})
```

```python
data = data.fillna(method='ffill', axis=1).fillna(method='bfill', axis=1).dropna(how='all')
```

```python
data['Sum emissions'] = data.sum(axis=1)
```

```python
data = data['Sum emissions']
```
