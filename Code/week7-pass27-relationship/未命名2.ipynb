```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as np
```

```python
df_temperature = pd.read_excel('GlobalTemperature.xlsx')
df_climate = pd.read_excel('ClimateChange.xlsx')
```

```python
df_temperature.Date = pd.to_datetime(df_temperature.Date)
```

```python
df_temperature_date = df_temperature.iloc[2880:3132,:]
```

```python
df_temperature_date
```

```python
df_temperature_result = df_temperature_date
df_temperature_result.index = df_temperature_date.Date
```

```python
df_temperature_resample = df_temperature_result.resample('A').sum()
```

```python
df_temperature_resample.head()
```

```python
df_temperature_resample1 = df_temperature_result.resample('A').mean()
```

```python
df_temperature_resample1
```

```python
df_temperature_drop = df_temperature_resample.drop(columns=
                                                   ['Land Max Temperature',
                                                    'Land Min Temperature'])
```

```python
df_temperature_drop.index = df_temperature_drop.index.year
```

```python
df_temperature_drop
```

```python
df_climate_series = df_climate.loc[df_climate['Series code'].isin (['EN.ATM.CO2E.KT','EN.ATM.METH.KT.CE',
                                       'EN.ATM.NOXE.KT.CE','EN.ATM.GHGO.KT.CE',
                                       'EN.CLC.GHGR.MT.CE' ])]
```

```python
df_climate_nan = df_climate_series.replace({'..':pd.np.NaN})
```

```python
df_climate_fill = df_climate_nan.iloc[:,6:27].
fillna(method='ffill',axis=1).fillna(method='bfill',axis=1)
```

```python
df_climate_drop = df_climate_fill.dropna(how='all').sum()
```

```python
df_climate_result = pd.DataFrame(df_climate_drop,columns=['co2-sum'])
```

```python
df_climate_result
```

```python
df_12 = pd.concat([df_climate_result,df_temperature_drop],axis=1)
```

```python
df_12
```

```python
df_12 = (df_12-df_12.min())/(df_12.max()-df_12.min())
```

```python
df_12
```
